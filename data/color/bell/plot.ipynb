```python
import glob

import matplotlib.pyplot as plt; plt.rcParams.update({"font.family": "serif"})
import numpy as np
```

```python
# Load experiment data matching these parameters.
computer: str = "fez"
```

```python
pattern = f"bell_steane_code_zz_ibm_{computer}_*"
```

```python
physical_values = []
encoded_values = []
encoded_dd_values = []
encoded_rem_values = []
encoded_dd_rem_values = []

for dir in glob.glob(pattern):
    physical_values.append(np.loadtxt(f"{dir}/physical.txt"))
    encoded_values.append(np.loadtxt(f"{dir}/encoded.txt"))
    encoded_dd_values.append(np.loadtxt(f"{dir}/encoded_dd.txt"))
    encoded_rem_values.append(np.loadtxt(f"{dir}/encoded_rem.txt"))
    encoded_dd_rem_values.append(np.loadtxt(f"{dir}/encoded_dd_rem.txt"))
```

```python
bins = 4
alpha = 0.75
edgecolor = "black"

plt.hist(physical_values, bins=bins, alpha=alpha, edgecolor=edgecolor, label="Physical")
plt.hist(encoded_values, bins=bins, alpha=alpha, edgecolor=edgecolor, label="Encoded")
# plt.hist(encoded_rem_values, bins=bins, alpha=alpha, edgecolor=edgecolor, label="Encoded + REM")
plt.hist(encoded_dd_values, bins=bins, alpha=alpha, edgecolor=edgecolor, label="Encoded + DD")
# plt.hist(encoded_dd_rem_values, bins=bins, alpha=alpha, edgecolor=edgecolor, label="Encoded + REM + DD")
plt.xlabel(r"$\langle \bar{Z} \bar{Z} \rangle$")
plt.title(f"Steane code Bell state preparation on IBM {computer.capitalize()}")
plt.axvline(1.0, color="xkcd:darkgreen", label="Ideal")
plt.legend();
```
